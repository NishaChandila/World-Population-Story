```python
import pandas as pd

file_path = "/content/drive/My Drive/Colab/csv_files/world_population.csv"
df = pd.read_csv(file_path)

df.head()
```

```python
df.columns
```

```python
columns = ["Continent", "Country/Territory", "1970 Population", "1980 Population", "1990 Population",
                   "2000 Population", "2010 Population", "2015 Population", "2020 Population", "2022 Population"]
df = df[columns]
df.head()
```

```python
df1 = df.melt(id_vars=["Continent", "Country/Territory"],
                   value_vars=["1970 Population", "1980 Population", "1990 Population",
                               "2000 Population", "2010 Population", "2015 Population", "2020 Population"],
                   var_name="Year",
                   value_name="Population")

df1["Year"] = df1["Year"].str.extract("(\d+)").astype(int)

df2 = df1.groupby(["Year", "Continent"])["Population"].sum().reset_index()

df2.head()
```

```python
df1.head()
```

```python
df1["Year"] = pd.to_datetime(df1["Year"], format="%Y")
drive_path = "/content/drive/My Drive/Colab/csv_files/world_population.csv"

df1.to_csv(drive_path, index=False)

print(f"File saved to: {drive_path}")
```
